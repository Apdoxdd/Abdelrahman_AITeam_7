```python
import numpy as np
import pandas as pd

df = pd.read_csv("housing.csv")

```

```python
print(df.shape)
print(df.head())
print(df.describe())
df.isnull().sum()
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
df = pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)
df['total_bedrooms'].isnull().sum()
```

set up the slicing part


```python
np.random.seed(12)
indices = np.random.permutation(len(df))
train_size = int(0.7 * len(df))
val_size   = int(0.15 * len(df))
train_idx = indices[:train_size]
val_idx = indices[train_size : train_size + val_size]
test_idx = indices[train_size + val_size:]
df_train = df.iloc[train_idx]
df_val = df.iloc[val_idx]
df_test = df.iloc[test_idx]
x_train = df_train.drop('median_house_value', axis=1).values.astype(float)
y_train = df_train['median_house_value'].values.reshape(-1, 1).astype(float)
x_test = df_test.drop('median_house_value', axis=1).values.astype(float)
y_test = df_test['median_house_value'].values.reshape(-1, 1).astype(float)
x_val = df_val.drop('median_house_value', axis=1).values.astype(float)
y_val = df_val['median_house_value'].values.reshape(-1, 1).astype(float)
```

```python
def add_bias_column(X):
    ones = np.ones((X.shape[0], 1))
    return np.concatenate((ones, X), axis=1)

x_train = add_bias_column(x_train)
x_val = add_bias_column(x_val)
x_test = add_bias_column(x_test)
```

now ill try to implement the normal values without anyscalling of values

```python
xT = x_train.T
xTx = xT @ x_train
xTx_inv = np.linalg.inv(xTx)
w = xTx_inv @ xT @ y_train
print("weights shape ", w.shape)
w[:5]
```

now we test the valutaion set


```python
y_pred_val = x_val @ w
y_pred_test = x_test @ w

mse_val = np.mean(np.square(y_val - y_pred_val))
mae_val = np.mean(np.abs(y_val - y_pred_val))
print(f"Validation MSE: {mse_val}")

print(f"Validation MAE: {mae_val}")
mse_test = np.mean(np.square(y_test - y_pred_test))
mae_test = np.mean(np.abs(y_test - y_pred_test))
print(f"test MSE: {mse_test}")

print(f"test MAE: {mae_test}")

```
