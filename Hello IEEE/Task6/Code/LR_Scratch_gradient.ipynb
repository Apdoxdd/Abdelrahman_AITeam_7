```python
import numpy as np
import pandas as pd

df = pd.read_csv("housing.csv")
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
df = pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)
df['total_bedrooms'].isnull().sum()
```

```python
np.random.seed(12)
indices = np.random.permutation(len(df))
train_size = int(0.7 * len(df))
val_size   = int(0.15 * len(df))
train_idx = indices[:train_size]
val_idx = indices[train_size : train_size + val_size]
test_idx = indices[train_size + val_size:]
df_train = df.iloc[train_idx]
df_val = df.iloc[val_idx]
df_test = df.iloc[test_idx]
x_train = df_train.drop('median_house_value', axis=1).values.astype(float)
y_train = df_train['median_house_value'].values.reshape(-1, 1).astype(float)
x_test = df_test.drop('median_house_value', axis=1).values.astype(float)
y_test = df_test['median_house_value'].values.reshape(-1, 1).astype(float)
x_val = df_val.drop('median_house_value', axis=1).values.astype(float)
y_val = df_val['median_house_value'].values.reshape(-1, 1).astype(float)


```

we need to scale down the values to a 0 - 1 range as not to have one feature over influences the others based on raw numbers

```python
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)
x_train_scaled = (x_train - mean) / std
x_val_scaled = (x_val - mean) / std
x_test_scaled = (x_test - mean) / std
```

```python
def add_bias(X):
    return np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)

X_train_final = add_bias(x_train_scaled)
X_val_final = add_bias(x_val_scaled)
X_test_final = add_bias(x_test_scaled)
```

```python
w_gd = np.zeros((X_train_final.shape[1], 1))

learning_rate = 0.01
epochs = 1000
n = len(y_train)
history = []

for i in range(epochs):
    y_hat = X_train_final @ w_gd
    gradient = (2/n) * X_train_final.T @ (y_hat - y_train)
    w_gd -= learning_rate * gradient
    mse = np.mean(np.square(y_hat - y_train))
    history.append(mse)
    if i % 100 == 0:
        print(f"epoch {i}: mSE = {mse}")
```
